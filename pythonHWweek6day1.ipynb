<h1>Voting Soft</h1>

```python
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
df.isnull().sum()
```

```python
model1 = SVC(probability=True)
model2 = LogisticRegression()
model3 = DecisionTreeClassifier()
```

```python
x = df.iloc[:, 0:7]
y = df.iloc[:,  -1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
voting_soft = VotingClassifier(estimators=[('svc', model1),
                                           ('lr', model2), 
                                           ('dt', model3)],
                               voting='soft')
```

```python
voting_soft.fit(x_train, y_train)
```

```python
voting_soft_pred = voting_soft.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, voting_soft_pred)
```

<h1>Voting Hard</h1>

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
x = df.iloc[:, 0:7]
y = df.iloc[:,  -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=101, test_size=0.3)
```

```python
model1 = SVC()
model2 = DecisionTreeClassifier()
model3 = LogisticRegression()
```

```python
voting_hard = VotingClassifier(estimators=[('svm', model1), ('dt', model2), ('lr', model3)], voting='hard')
```

```python
voting_hard.fit(x_train, y_train)
```

```python
voting_hard_pred = voting_hard.predict(x_test)
```

```python
accuracy_score(y_test, voting_hard_pred)
```

<h1>Bagging Classifier</h1>

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
estimators = SVC(probability=True)
bagging_classifier = BaggingClassifier(estimator=estimators, n_estimators=10, max_samples=0.4,
                        bootstrap=True, bootstrap_features=True, n_jobs=-1, random_state=123)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=101, test_size=0.3)
```

```python
bagging_classifier.fit(x_train, y_train)
```

```python
bagging_pred = bagging_classifier.predict(x_test)
```

```python
accuracy_score(y_test, bagging_pred)
```

<h1>Gradient Boosting Classifier</h1>

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
gbc_pred = gbc.predict(x_test)
```

```python
accuracy_score(gbc_pred, y_test)
```
